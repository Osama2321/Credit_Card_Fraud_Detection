```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("creditcard.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
# 0 represents normal transactions
# 1 reprsents fraud transactions
df['Class'].value_counts()
```

```python
#separating data for analysis
normal=df[df.Class==0]
fraud=df[df.Class==1]
```

```python
print(normal.shape)
print(fraud.shape)
```

```python
#statistics measure of the data
normal.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
df.groupby('Class').mean()
```

```python
legit_sample=normal.sample(n=492)
```

```python
new_dataset=pd.concat([legit_sample,fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python



X=new_dataset.drop(columns='Class',axis=1)
y=new_dataset['Class']
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
```

```python
#Splitting Data into training and testing
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
#Applying model
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
#Evaluation
training_data_accuracy=accuracy_score(model.predict(X_train),y_train)
training_data_accuracy
```

```python
testing_data_accuracy=accuracy_score(model.predict(X_test),y_test)
testing_data_accuracy
```
